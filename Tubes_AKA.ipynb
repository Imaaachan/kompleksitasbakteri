```python
import math
import numpy as np

def pertumbuhan_bakteri_3d(waktu_total):
    """
    Simulasi pertumbuhan bakteri dalam ruang 3 dimensi.
    
    Args:
    waktu_total: int, waktu total dalam jam.
    """
    # Parameter tetap
    grid_size = (2, 2, 2)  # Ukuran grid 3D (x, y, z)
    n0 = 1.0               # Populasi awal per sel
    k = 0.1                # Laju pertumbuhan
    interval = 1           # Interval utama (dalam jam)
    langkah_kecil = 1     # Langkah waktu kecil (dalam jam)
    
    print("Waktu\tTotal Populasi")
    
    # Inisialisasi grid 3D dengan populasi awal
    grid = np.full(grid_size, n0, dtype=float)
    waktu = 0
    
    # Iterasi waktu
    while waktu <= waktu_total:
        # Nested loop untuk langkah kecil
        for _ in range(0, interval, langkah_kecil):
            # Pertumbuhan setiap sel di grid
            grid *= math.exp(k * langkah_kecil)
        
        # Output total populasi di grid pada interval utama
        total_populasi = np.sum(grid)
        print(f"{waktu}\t{round(total_populasi, 2)}")
        waktu += interval

# Input parameter simulasi
waktu_total = int(input("Masukkan waktu total (dalam jam): "))

# Jalankan simulasi
pertumbuhan_bakteri_3d(waktu_total)
```
